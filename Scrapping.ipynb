```python
from bs4 import BeautifulSoup
from urllib.request import urlopen as ureq
import time
import os
import requests
import numpy as np
import pandas as pd
```

```python
from selenium import webdriver
import re
```

```python
browser = webdriver.Chrome()
browser.get("https://www.nobroker.in/")
```

```python
link=[]
containers = browser.find_elements_by_id("listCardContainer")

for i in containers:
    links = i.find_elements_by_class_name("nb__sDz3z")
    for j in links:
        lnks=j.find_elements_by_tag_name("a")
        link.append(lnks[0].get_attribute('href'))
```

```python
len(link)
```

```python
link
```

```python
rent_attr=[]
for i in link:
    url = requests.get(i).text
    page_soup_2 = BeautifulSoup(url,'lxml')

    name = page_soup_2.find_all("div",{"class":"nb__3V5eg"})[0].contents[0]
    name= name.text.strip()


    rent = page_soup_2.find_all("div",{"class":"nb__Qljzo"})[0].contents[0]
    rent = rent.text.strip()
    num_rent = ""
    for i in range(len(rent)):
        if (rent[i].isdigit()):
            num_rent = num_rent + rent[i]


    area = page_soup_2.find_all("div",{"class":"nb__Og576"})[0]
    area = area.text.strip()
    area = area[:-5]

    deposit = page_soup_2.find_all("div",{"class":"nb__3zfm-"})[0]
    deposit = deposit.text.strip()
    deposit = deposit[2:]
    deposit = deposit.replace("Deposit","")

    furnishing = page_soup_2.find_all("div",{"class":"nb__2xbus"})[0].contents[0]
    furnishing = furnishing.text.strip()

    facing = page_soup_2.find_all("div",{"class":"nb__2xbus"})[1].contents[0]
    facing = facing.text.strip()

    water = page_soup_2.find_all("div",{"class":"nb__2xbus"})[2].contents[0]
    water = water.text.strip()

    floor = page_soup_2.find_all("div",{"class":"nb__2xbus"})[3].contents[0]
    floor = floor.text.strip()

    bathroom = page_soup_2.find_all("div",{"class":"nb__2xbus"})[4].contents[0]
    bathroom = bathroom.text.strip()

    security = page_soup_2.find_all("div",{"class":"nb__2xbus"})[6].contents[0]
    security = security.text.strip()

    bedroom = page_soup_2.find_all("div",{"class":"nb__3vD7l"})[0].contents[0]
    bedroom = bedroom.text.strip()
    bedroom = bedroom.replace("Bedroom","")

    apt_type = page_soup_2.find_all("div",{"class":"nb__3vD7l"})[1].contents[0]
    apt_type = apt_type.text.strip()

    parking = page_soup_2.find_all("div",{"class":"nb__3vD7l"})[4].contents[0]
    parking=parking.text.strip()

    age_building = page_soup_2.find_all("div",{"class":"nb__3vD7l"})[5].contents[0]
    age_building=age_building.text.strip()

    balcony = page_soup_2.find_all("div",{"class":"nb__3vD7l"})[6].contents[0]
    balcony = balcony.text.strip()

    house_rent={'Name':name,'Rent':num_rent,"Area":area,"Deposit":deposit,"Furnishing":furnishing,"Facing": facing,"Water_Supply":water,
               "Floor":floor,"Bathroom":bathroom,"Security":security,"BHK":bedroom,"Type":apt_type,"Parking":parking,"Age_Building":age_building,
               "Balcony":balcony}
    rent_attr.append(house_rent)



```

```python
df_gurgaon= pd.DataFrame(rent_attr)
df_gurgaon
```

```python
df_gurgaon.to_csv("new_delhi.csv",index=False)
```
